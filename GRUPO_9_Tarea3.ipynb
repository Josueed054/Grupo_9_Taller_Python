```python
import scipy.stats as stats

# Parámetros de la distribución binomial
n = 10  # Número de lanzamientos
p = 0.5  # Probabilidad de obtener una cara en cada lanzamiento
k = 7    # Número de caras que queremos obtener

# Calcular la probabilidad de obtener exactamente 7 caras en 10 lanzamientos
probabilidad = stats.binom.pmf(k, n, p)

print(f"La probabilidad de obtener {k} caras en {n} lanzamientos es {probabilidad}")
```

```python
#Probabilidad acumulada de obtener menos de 7 caras en 10 lanzamientos
probabilidad_acumulada = stats.binom.cdf(k, n, p)

print(f"La probabilidad acumulada de obtener menos de {k} caras en {n} lanzamientos es {probabilidad_acumulada}")
```

```python
#Creo una lista del rango de número de intentos
número_de_intentos = list(range(1, 11))
print(número_de_intentos)
#Evaluación de cada caso
for x in número_de_intentos:
      probabilidad_1 = stats.binom.pmf(x, n, p)
      print(probabilidad_1)
plt.bar(número_de_intentos, probabilidades)
plt.xlabel('Número de caras')
plt.ylabel('Probabilidad')
plt.title('Distribución binomial')
plt.show()
```

```python


# Parámetros de la distribución
n = 10
p_2 = 0.8

# Se crea una lista con el rango de intentos
número_de_intentos = list(range(0, n + 1))

# Evaluación de cada caso
probabilidades = [stats.binom.pmf(x, n, p_2) for x in número_de_intentos]

# Graficar la distribución binomial
plt.bar(número_de_intentos, probabilidades)
plt.xlabel('Número de caras')
plt.ylabel('Probabilidad')
plt.title('Distribución binomial')
plt.show()
```

```python
# Parámetros de la distribución normal
media = 0
desviacion_estandar = 1

# Generar 1000 muestras aleatorias
datos = stats.norm.rvs(loc=media, scale=desviacion_estandar, size=1000)
```

```python
# Creamos el histograma
plt.hist(datos, bins=30, density=True, alpha=0.6, color='r')

# Se agrega la curva a la distribución normal
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, media, desviacion_estandar)
plt.plot(x, p, 'k', linewidth=2)
plt.title("Distribución Normal (Media = 0, Desviación Estándar = 1)")
plt.xlabel("Valor")
plt.ylabel("Densidad de probabilidad")
plt.show()
```

```python

# Calcular la probabilidad acumulada de -1 a 1
probabilidad = stats.norm.cdf(1, loc=media, scale=desviacion_estandar) - stats.norm.cdf(-1, loc=media, scale=desviacion_estandar)

print(f"La probabilidad con el rango de -1 a 1 es: {probabilidad}")
```
